Load the querries

```python
queries = dict()
with open("./output/queries.txt", "r") as f:
    for line in f:
        cols = line.split("\t")
        queries[cols[0].strip()] = cols[1].strip()
```

Implement the Retrieval Model:

```python
import math
from pyserini.index import IndexReader
import numpy as np
index_path = './output/indexes/lucene_index_json'

index_reader = IndexReader(index_path)
len_C = index_reader.stats()["total_terms"]

def len_doc(d):
    len_d = sum([v for k,v in index_reader.get_document_vector(d).items()])
    return len_d

def coll_freq(t):
    cf = index_reader.get_term_counts(t)[1]
    return cf

def term_freq(t, d):
    dv = index_reader.get_document_vector(d)
    tf = 0
    if t in dv: tf = dv[t]
    return tf

def prob_t_Md(t, d, lambd):
    a = (1-lambd)*(term_freq(t, d)/len_doc(d))
    b = lambd*(coll_freq(t)/len_C)
    p_t_Md = np.log(a+b)
    return p_t_Md

def score_doc(q, d, lambd):
    p_q_Md = 0
    for t in q:
        if coll_freq(t) != 0: 
            p_q_Md += prob_t_Md(t, d, lambd)
    return p_q_Md
```

Perform the Retrieval based on q querries

```python
from pyserini.search import SimpleSearcher

searcher = SimpleSearcher(index_path)
BM25 = True
n_hits = 100

results = []
for q_id in queries:
    hits = searcher.search(queries[q_id], 100)
    for i in range(0, min(len(hits), n_hits)):
        d = hits[i].docid
        if BM25: score = hits[i].score
        else:
            query = index_reader.analyze(queries[q_id])
            score = score_doc(query, d, 0.1)

        results.append([q_id, d, score])
    
check = set()
for res in results:
    if ((res[0], res[1])) in check:
        raise Exception("Error: Duplicate query-doc is found", res[0], res[1])
    check.add((res[0], res[1]))
    
output_str = "\n".join([l[0] + "\t" + l[1] + "\t" + str(l[2]) for l in results])
open("./output/pyserini.run", "w").write(output_str)

#https://github.com/castorini/pyserini/blob/master/docs/experiments-robust04.md
#!python -m pyserini.search --topics robust04 --index robust04 --output run.robust04.txt --bm25
#!python -m pyserini.eval.trec_eval -m map -m P.30 robust04 run.robust04.txt
```
